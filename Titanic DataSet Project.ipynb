```python
#Who was alone and who was with family
titanic_df['Alone'] = titanic_df.SibSp + titanic_df.Parch
```

```python
def aloneOrFamily(relativeInfo):
    sib,parch = relativeInfo
    if sib+parch != 0:
            Alone = 'With Family'
    else:
            Alone = 'Without Family'
    return Alone
```

```python
sns.factorplot('Alone',kind = 'count',data=titanic_df)
```

```python
titanic_df['Alone'] = titanic_df[['SibSp','Parch']].apply(aloneOrFamily,axis=1)
```

```python
#What factor helps the people in surviving the titanic crash
titanic_df['Survivor'] = titanic_df.Survived.map({0:'no',1:'yes'})
```

```python
class deckInfo():
    def __init__(self):
        self.deck = [] 
        self.survived = []
        
def deckWiseSurvival(Cabin,Survived):
    dI = deckInfo()
    for i in range(len(Survived)):
        if Cabin[i] != 'NaN':
            try:
                dI.deck.append(str(Cabin[i])[0])
                dI.survived.append(Survived[i])
            except:
                dI.deck = str(Cabin[i])[0]
                dI.survived = [Survived[i]] 
    print(len(dI.deck))
    print(len(dI.survived))
    return dI
```

```python
deckSurvivorInfo= deckWiseSurvival(titanic_df['Cabin'],titanic_df['Survived'])
```

```python
#List of tuples
deckSurvivor_df = DataFrame( {
                              'deck':deckSurvivorInfo.deck,
                              'Survivor':deckSurvivorInfo.survived
                             } )
```
